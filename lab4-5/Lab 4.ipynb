Save a copy of the original data

```python
old_data = data.copy()
old_data.head()
```

Select the data to be working with


```python
old_short_data = old_data [:120].copy()
old_short_data = old_short_data.values
new_data = data [:120].copy()
new_data = new_data.values
```

```python
print(new_data)
```

Label 10% of the data cells as 99

```python
#Label 10% of the dataset cells as 99


amount = 252318
a =0
count = 0
for i in range(0,100):
        for j in range(0,120):
            a = a + 1
            if count < amount:
                if a == 7:
                    a = 0
                    new_data[j][i] = 99
print(new_data)

```

Prepare the information for the training

```python
n_features = 2

user_ratings = new_data
latent_user_preferences = np.random.random((user_ratings.shape[0], n_features))
latent_item_features = np.random.random((user_ratings.shape[1],n_features))
```

```python
def predict_rating(user_id,item_id):
    """ Predict a rating given a user_id and an item_id.
    """
    user_preference = latent_user_preferences[user_id]
    item_preference = latent_item_features[item_id]
    return user_preference.dot(item_preference)

def train(user_id, item_id, rating,alpha = 0.001):
    prediction_rating = predict_rating(user_id, item_id)
    err =  ( prediction_rating - rating );
    user_pref_values = latent_user_preferences[user_id][:]
    latent_user_preferences[user_id] -= alpha * err * latent_item_features[item_id]
    latent_item_features[item_id] -= alpha * err * user_pref_values
    return err

def test(user_id, item_id, rating,alpha = 0.001):
    prediction_rating = predict_rating(user_id, item_id)
    err_test =  ( prediction_rating - rating );
    return err_test



def sgd(iterations = 3000):
    """ Iterate over all users and all items and train for 
        a certain number of iterations
    """
    for iteration in range(0,iterations):
        error = []
        error_test=[]
        for user_id in range(0,latent_user_preferences.shape[0]):
            for item_id in range(0,latent_item_features.shape[0]):
                rating = user_ratings[user_id][item_id]
                rating_test = old_short_data[user_id][item_id]
                # We check if the value is not 99
                if(rating != 99 ):
                    err = train(user_id,item_id,rating)
                    error.append(err)
                if(rating_test != 99 ):
                    err_test = test(user_id,item_id,rating_test)
                    error_test.append(err_test)
        mse = (np.array(error) ** 2).mean()
        mse_test = (np.array(error_test) ** 2).mean()
        if(iteration%100 == 0 ):
            print ("mse in the train iteration",iteration,"is",mse)
            print ("mse in the test iteration",iteration,"is",mse_test)
    print("Final train mse", mse)
    print("Final test mse", mse_test)
```

```python
sgd()
```
